### 3. `sorted() vs .sort()`

전자: sort list and **_RETURN_**  
후자: sort list **_INPLACE_**

```python
a = [1, 3, 2]
b = sorted(a)
c = [1, 3, 2]
c.sort()

print(a) # doesn't changed
print(b) # return sorted "a"
print(c) # changed
```

### 내 풀이

```python
def solution(array, commands):
    answer = []
    for i, j, k in commands:
        # 잘라서 정렬
        temp = sorted(array[i-1:j])
        # k번째 answer에 저장
        answer.append(temp[k-1])
    
    return answer
```

#### 재웅님

```python
def solution(array, commands):
    answer = []
    
    
    for i in range(len(commands)):
        # commands에서 0번과 1번을 꺼내서 인덱스 슬라이싱을 해줍니다. 단, 시작지점에서 1을 빼주어야 원하는 값을 가져올 수 있습니다.(배열의 특성상 0에서 시작하므로)
        new = array[commands[i][0] - 1: commands[i][1]]
        # 정렬
        new.sort()
        # 위와 마찬가지로 1을 빼주어서 실제로 원하는 위치의 값을 가져옵니다.
        answer.append(new[commands[i][2] - 1])
    
    return answer
```

#### 하정님

```python
def solution(array, commands):
    answer = []
    
    # 문제 설명에서 설명해준 것과 같은 순서로 진행
    # 1. 인덱스를 이용해 리스트 슬라이싱
    # 2. 슬라이싱한 리스트를 정렬
    # 3. 그 중 k번째 수 정답 리스트에 넣기
    for i in commands:
        cnt = 1
        for j in sorted(array[i[0]-1:i[1]]):
            if cnt == i[2]:
                answer.append(j)
                break
            cnt += 1
        # 필요 없는 예외처리
        else : answer.append(-1)
    return answer
```

## 2) 가장 큰 수

아이디어: 왜 10보다 2가 먼저 왔을까?  
$\rightarrow$ 길이에 상관 없이 **앞에서부터 읽어서 크기 비교**  
$\quad$`numbers.sort(key=lambda x: str(x)[0])`  

$\rightarrow$ 근데 각 원소는 최대 세 자리(1000은 비교 대상 없음)    
$\quad$`numbers.sort(key=lambda x: (str(x)[0], str(x)[1], str(x)[2]))`  

$\rightarrow$ `IndexError` 발생  
$\Rightarrow$ **모든 문자열의 길이를 늘리자**

### 내 풀이

```python
def solution(numbers):
    # 반복과 join을 위해 str 배열로 변경
    numbers = list(map(str, numbers))
    # 세 번 반복해서 큰 순으로 정렬
    # -> 문자열은 자리수에 관계 없이 앞부터 각각 비교한다.
    numbers.sort(key=lambda x: x * 3, reverse=True)
    
    # join으로 정답 문자열 생성
    answer = ''.join(numbers)
    # 맨 앞이 0이다 == 애초에 0만 들어왔다
    # -> 답은 '0'
    if answer[0] == '0':
        answer = '0'
    
    return answer
```

#### 재석님

```python
def solution(numbers):
    answer = ''
    # 숫자들을 이어붙여야 하므로 우선 numbers를 list 형태로 만듬
    numbers = list(map(str, numbers))
    # 같은 숫자로 연속했을 때 앞에 오도록 정렬 ex) 90보다 9가 앞에 있는 것이 더 큰 수를 만들 수 있기 때문
    numbers.sort(key=lambda x: x * 5, reverse=True)    # 굳이 5번 할 필욘 없음 / 띄어쓰기x
    # 정렬된 숫자들로 이어붙임
    for i in numbers:
        answer += i
    # 000.. 같은 테스트 케이스가 있기 때문에 answer를 int형으로 만들었다가 다시 str형태로 
    return str(int(answer))                                # 이렇게 하면 느림
```

```python
def solution(numbers):
    answer = ''
    
    # numbers 리스트를 스트링 리스트로 바꿔준다
    numbers = list(map(str, numbers))
    # 가장 큰 수를 찾기 위해 여러 번 반복한 형태로 sorting을 해준다 
    # ex) 34일때 343434 30일때 303030
    numbers.sort(key= lambda x: x*3, reverse=True)
    
    # sorting 된 리스트를 스트링으로 이어준다
    answer = ''.join(numbers)
    
    # 0만 여러개 나왔을 경우 '0000'이 아닌 '0'이 출력될 수 있도록 처리   
    return '0' if not int(answer) else answer    # 재석님보단 빠름
```

#### 재웅님

```python
# 타임에러...for문 두개라 그런듯합니다...
import itertools # 숫자들의 조합을 위해 itertools 사용
def solution(numbers):
    answer = ''
    temp = ''
    numbers = list(map(str, numbers)) # 숫자들을 문자형으로 변형시킴
    result = list(itertools.permutations(numbers, len(numbers))) # 조합들이 저장된 리스트
    temp_list = []

    for i in result: # 순열들 내부의 값들을 합해줌
        for j in range(len(i)): 
            temp = temp + i[j]   # temp += i[j]
        temp_list.append(temp)    
        temp = ''        

    '''max를 사용하면 됩니다'''
    temp_list.sort(reverse=True) # 더해준것이 저장된 리스트를 역으로 sort해주어 최대값을 찾아줌
    answer = temp_list[0]
        
        
    return answer
```

#### 의현님

```python
# 타임에러 permutation 에서 생기는거 같습니다
from itertools import permutations

def solution(numbers):
    # 순열만들고
    # 맥스 리턴
    permute = list(permutations(numbers, len(numbers)))
    '''재웅님의 이중 for문 구현 (map == for)'''
    list_permute = [''.join(map(str, i)) for i in permute]
    answer = max(list_permute)
    
    return answer
```

#### 영준님

```python
# 풀이 실패 ㅠㅠㅠ
# 테스트케이스는 다 맞으나 실제로 돌렸을때 오류 ㅠ

def solution(numbers):
    # numbers들의 숫자 원소들을 문자로 변환하여 일차적으로 정렬함
    str_numbers = []
    for num in numbers:
        str_numbers.append(str(num))

    str_numbers.sort(reverse = True)
    
    # 결국 앞자리가 다른 원소들은 우선적으로 정렬
    # 그러나 앞자리가 같은 원소들끼리의 정렬이 문제임
    # for문으로 순환하면서
    for i in range(len(str_numbers) - 1):
        # i번째와 i + 1번째 원소가 앞자리가 같을 경우,
        if str_numbers[i][0] == str_numbers[i + 1][0]:
            # 두 문자열을 합한 것을 int했을 때, 순서를 바꾼 것이 안바꾼 것보다 더 크면
            if int(str_numbers[i] + str_numbers[i + 1]) < int(str_numbers[i + 1] + str_numbers[i]):
                # 순서를 바꿔준다.
                str_numbers[i], str_numbers[i + 1] = str_numbers[i + 1], str_numbers[i]
                
    # 만약 [0,0,0,0] 등 0으로만 이루어진 문자열일 경우
    if str_numbers[0] == '0':
        answer = '0'
    else:
        answer = "".join(str_numbers)
    
    return answer
```
